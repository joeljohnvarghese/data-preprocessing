```python
import numpy as np
import pandas as pd
```

```python
#duplicate
dataset1=pd.read_csv("/content/sample_data/diabetes.csv")
```

```python
dataset1
```

```python
#categorical
dataset1["Outcome"]=dataset1["Outcome"].apply(lambda e:0 if e=="N" else 1)
```

```python
dataset1
```

```python
#x,y=dataset1.drop(["Outcome"],axis=1),dataset["Outcome"]
```

```python
# Check if there are any duplicates
dataset1.duplicated()
```

```python
#removal
dataset1.drop_duplicates(inplace=True)
dataset1
```

```python
# Check if there is any missing data
dataset1.isnull()
```

```python
# Fill missing data with 0
dataset1.fillna(0)
```

```python
# Or Drop entries with missing data
dataset1.dropna()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.NaN, strategy='mean')
imputer.fit(dataset1["Age"].values.reshape(-1,1))
```

```python
dataset1["Age"]=imputer.transform(dataset1["Age"].values.reshape(-1,1))
```

```python
imputer=SimpleImputer(missing_values=np.NaN, strategy='mean')
imputer.fit(dataset1["Glucose"].values.reshape(-1,1))
dataset1["Glucose"]=imputer.transform(dataset1["Glucose"].values.reshape(-1,1))
```

```python
imputer=SimpleImputer(missing_values=np.NaN, strategy='mean')
imputer.fit(dataset1["BloodPressure"].values.reshape(-1,1))
dataset1["BloodPressure"]=imputer.transform(dataset1["BloodPressure"].values.reshape(-1,1))
```

```python
imputer=SimpleImputer(missing_values=np.NaN, strategy='mean')
imputer.fit(dataset1["SkinThickness"].values.reshape(-1,1))
dataset1["SkinThickness"]=imputer.transform(dataset1["SkinThickness"].values.reshape(-1,1))
```

```python
imputer=SimpleImputer(missing_values=np.NaN, strategy='mean')
imputer.fit(dataset1["Insulin"].values.reshape(-1,1))
dataset1["Insulin"]=imputer.transform(dataset1["Insulin"].values.reshape(-1,1))
```

```python
imputer=SimpleImputer(missing_values=np.NaN, strategy='most_frequent')
imputer.fit(dataset1["Pregnancies"].values.reshape(-1,1))
dataset1["Pregnancies"]=imputer.transform(dataset1["Pregnancies"].values.reshape(-1,1))
```
